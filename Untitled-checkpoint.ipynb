```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay
)

# Set a random seed so results are reproducible (same results every run)
np.random.seed(42)

# Make our plots look clean and professional
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

print("✅ All libraries imported successfully!")
```

```python
n_samples = 1000  # Total number of records (days) to simulate

# Temperature in Celsius: ranging from 15°C (cool) to 45°C (very hot)
temperature = np.random.uniform(15, 45, n_samples)

# Relative Humidity in %: ranging from 10% (very dry) to 90% (very humid)
humidity = np.random.uniform(10, 90, n_samples)

# Wind speed in km/h: ranging from 0 (calm) to 60 (strong winds)
wind = np.random.uniform(0, 60, n_samples)

# ─────────────────────────────────────────────────────────────────
# Create the target label: fire (1 = fire occurred, 0 = no fire)
#
# We compute a "fire score" based on weather conditions:
#   - Higher temperature  → increases fire risk
#   - Higher humidity     → decreases fire risk
#   - Higher wind speed   → increases fire risk
# A random noise term is added to make it realistic (not perfectly predictable)
# ─────────────────────────────────────────────────────────────────

fire_score = (
    0.05 * temperature       # Temperature increases risk
    - 0.04 * humidity        # Humidity decreases risk
    + 0.03 * wind            # Wind increases risk
    + np.random.normal(0, 0.5, n_samples)  # Add some random noise
)

# Convert the continuous score into a binary label:
# If fire_score > 1.0, we label it as fire (1), otherwise no fire (0)
fire = (fire_score > 1.0).astype(int)

# ─────────────────────────────────────────────────────────────────
# Assemble everything into a pandas DataFrame (like a spreadsheet)
# ─────────────────────────────────────────────────────────────────
df = pd.DataFrame({
    "temperature": temperature.round(2),
    "humidity": humidity.round(2),
    "wind": wind.round(2),
    "fire": fire
})

print(f"✅ Dataset created with {len(df)} rows and {len(df.columns)} columns.")
print(f"   Fire cases: {fire.sum()} | No-fire cases: {(fire == 0).sum()}")
```

```python
print("📋 First 5 rows of the dataset:")
df.head()
```

```python
print("📊 Statistical Summary:")
df.describe()
```

```python
print("🔍 Missing values per column:")
print(df.isnull().sum())
print()

# Check data types of each column
print("📐 Data types:")
print(df.dtypes)
```

```python
print("🔥 Class Distribution (0 = No Fire, 1 = Fire):")
print(df["fire"].value_counts())
print()
print(f"   Fire rate: {df['fire'].mean()*100:.1f}%")

# Visualize the class distribution
fig, ax = plt.subplots()
df["fire"].value_counts().plot(
    kind="bar", ax=ax,
    color=["#3498db", "#e74c3c"],
    edgecolor="black"
)
ax.set_title("Class Distribution: Fire vs No Fire", fontsize=14, fontweight="bold")
ax.set_xlabel("Fire Label (0 = No Fire, 1 = Fire)", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.set_xticklabels(["No Fire", "Fire"], rotation=0)
plt.tight_layout()
plt.show()
```

```python
initial_count = len(df)
df = df.drop_duplicates()
print(f"🧹 Removed {initial_count - len(df)} duplicate rows.")

# Step 2: Drop rows with any missing values
df = df.dropna()
print(f"🧹 Rows after dropping NaN: {len(df)}")

# Step 3: Validate value ranges (simple sanity check)
assert df["temperature"].between(0, 60).all(), "Temperature out of range!"
assert df["humidity"].between(0, 100).all(), "Humidity out of range!"
assert df["wind"].between(0, 150).all(), "Wind out of range!"
assert df["fire"].isin([0, 1]).all(), "Fire label should be 0 or 1!"

print("✅ Data cleaning complete. All checks passed!")
print(f"   Final dataset size: {len(df)} rows")
```

```python
features = ["temperature", "humidity", "wind"]
feature_labels = ["Temperature (°C)", "Humidity (%)", "Wind Speed (km/h)"]
colors = {0: "#3498db", 1: "#e74c3c"}

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Feature Distributions: Fire vs No Fire", fontsize=15, fontweight="bold")

for ax, feature, label in zip(axes, features, feature_labels):
    for fire_val, group in df.groupby("fire"):
        ax.hist(
            group[feature],
            bins=25,
            alpha=0.6,
            color=colors[fire_val],
            label="Fire" if fire_val == 1 else "No Fire",
            edgecolor="white"
        )
    ax.set_title(label, fontsize=13, fontweight="bold")
    ax.set_xlabel(label)
    ax.set_ylabel("Count")
    ax.legend()

plt.tight_layout()
plt.show()

# ─────────────────────────────────────────────────────────────────
# Observation:
#   - Fire tends to occur at HIGHER temperatures
#   - Fire tends to occur at LOWER humidity
#   - Fire tends to occur at HIGHER wind speeds
# ─────────────────────────────────────────────────────────────────
print("📌 Observation: Notice how fire cases (red) cluster at high temp, low humidity, high wind!")
```

```python
X = df[["temperature", "humidity", "wind"]]

# Target (output): fire label (0 or 1)
y = df["fire"]

print(f"📐 Feature matrix X shape: {X.shape}  (rows × features)")
print(f"📐 Target vector y shape:  {y.shape}")
print()
print("Sample of X (features):")
print(X.head(3))
print()
print("Sample of y (labels):")
print(y.head(3).values)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,     # 20% for testing, 80% for training
    random_state=42,   # Ensures the same split every time we run
    stratify=y         # Maintain class balance in both splits
)

print(f"✅ Data split complete!")
print(f"   Training samples : {len(X_train)}")
print(f"   Testing  samples : {len(X_test)}")
print()
print(f"   Fire rate in training set : {y_train.mean()*100:.1f}%")
print(f"   Fire rate in test set     : {y_test.mean()*100:.1f}%")
```

```python
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=6,
    random_state=42
)

# ─────────────────────────────────────────────────────────────────
# Train ("fit") the model on the training data.
# The model learns patterns from X_train to predict y_train.
# ─────────────────────────────────────────────────────────────────

model.fit(X_train, y_train)

print("✅ Model training complete!")
print(f"   Trees in the forest : {model.n_estimators}")
print(f"   Max depth per tree  : {model.max_depth}")
print(f"   Features used       : {list(X.columns)}")
```

```python
y_pred = model.predict(X_test)

# ─────────────────────────────────────────────────────────────────
# Accuracy Score:
# The percentage of predictions that were correct.
# Formula: correct predictions / total predictions
# ─────────────────────────────────────────────────────────────────

accuracy = accuracy_score(y_test, y_pred)
print(f"🎯 Model Accuracy: {accuracy * 100:.2f}%")
print()
print("Interpretation:")
if accuracy >= 0.90:
    print("  ✅ Excellent! The model is highly accurate.")
elif accuracy >= 0.80:
    print("  ✅ Good accuracy. The model is performing well.")
else:
    print("  ⚠️ Moderate accuracy. Consider adding more features.")
```

```python
cm = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(6, 5))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["No Fire", "Fire"]
)
disp.plot(ax=ax, cmap="Reds", colorbar=False)

ax.set_title("Confusion Matrix", fontsize=14, fontweight="bold", pad=15)
ax.set_xlabel("Predicted Label", fontsize=12)
ax.set_ylabel("Actual Label", fontsize=12)
plt.tight_layout()
plt.show()

tn, fp, fn, tp = cm.ravel()
print(f"True Positives  (Fires correctly detected)  : {tp}")
print(f"True Negatives  (No-fires correctly detected): {tn}")
print(f"False Positives (False Alarms)               : {fp}")
print(f"False Negatives (Missed Fires!) ⚠️            : {fn}")
```

```python
print("📋 Classification Report:")
print("─" * 55)
print(classification_report(y_test, y_pred, target_names=["No Fire", "Fire"]))
```

```python
importances = model.feature_importances_
feature_names = X.columns.tolist()

# Put importances in a DataFrame and sort from highest to lowest
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances
}).sort_values("Importance", ascending=True)  # ascending for horizontal bar chart

# Create a color-coded bar chart
bar_colors = ["#e74c3c" if imp == importance_df["Importance"].max() else "#3498db"
              for imp in importance_df["Importance"]]

fig, ax = plt.subplots(figsize=(8, 4))

bars = ax.barh(
    importance_df["Feature"],
    importance_df["Importance"],
    color=bar_colors,
    edgecolor="black",
    height=0.5
)

# Add value labels on each bar
for bar, val in zip(bars, importance_df["Importance"]):
    ax.text(
        bar.get_width() + 0.003,
        bar.get_y() + bar.get_height() / 2,
        f"{val:.3f}",
        va="center", fontsize=11, fontweight="bold"
    )

ax.set_title(" Feature Importance (Random Forest)", fontsize=14, fontweight="bold")
ax.set_xlabel("Importance Score", fontsize=12)
ax.set_ylabel("Weather Feature", fontsize=12)
ax.set_xlim(0, importance_df["Importance"].max() + 0.08)
ax.yaxis.set_tick_params(labelsize=12)

plt.tight_layout()
plt.show()

# Print a readable summary
print("\n Feature Importance Ranking:")
for _, row in importance_df.sort_values("Importance", ascending=False).iterrows():
    bar = "█" * int(row["Importance"] * 50)
    print(f"   {row['Feature']:15s} | {bar} {row['Importance']:.3f}")
```

```python
custom_scenarios = pd.DataFrame({
    "temperature": [42,  18,  38,  25],   # Degrees Celsius
    "humidity":    [15,  80,  20,  55],   # Percent
    "wind":        [45,   5,  30,  10],   # km/h
})

scenario_labels = [
    "Extreme heat, dry, strong wind",
    "Cool, humid, calm",
    "Hot, dry, moderate wind",
    "Mild, moderate humidity, calm"
]

# Get predictions and probabilities
predictions = model.predict(custom_scenarios)
probabilities = model.predict_proba(custom_scenarios)[:, 1]  # Probability of fire

print("🔥 Custom Prediction Results:")
print("─" * 70)
print(f"{'Scenario':<40} {'Prediction':<15} {'Fire Prob.':<10}")
print("─" * 70)
for label, pred, prob in zip(scenario_labels, predictions, probabilities):
    result = "🔥 FIRE" if pred == 1 else "✅ NO FIRE"
    print(f"{label:<40} {result:<15} {prob*100:.1f}%")
```
